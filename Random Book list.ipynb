```python
#DEPENDENCIES
import os
import pandas as pd
import random

bookDataLocation = "resources/books.csv"
```

```python
df = pd.read_csv(bookDataLocation)

df.head(10)
```

```python
ISBN_list = pd.DataFrame(columns=["title", "authors", "isbn"])

for x in range(100):
    randomBook = random.randint(1,10000)
    book = df.loc[df["bookID"] == randomBook, ["title", "authors", "isbn"]]
    ISBN_list = ISBN_list.append(book)
    
ISBN_list.head()
```
