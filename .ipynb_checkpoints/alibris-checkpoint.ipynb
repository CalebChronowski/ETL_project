```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

import os
import random


bookDataLocation = "resources/books.csv"
```

```python
df = pd.read_csv(bookDataLocation)

ISBN_list = pd.DataFrame(columns=["title", "authors", "isbn"])

def random_book():
    randomBook = random.randint(1,10000)
    book = df.loc[df["bookID"] == randomBook, ["title", "authors", "isbn"]]
    return(book)
    
random_book()
```

```python
#Links
isbns = random_book().isbn.tolist()
links=isbns
```

```python
#Extract Power tool data

executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)

alibris_pandas_df = pd.DataFrame([])

#This doesn't work right now because there isn't a list of links
for x in links:

    try:
        url = (f'https://www.alibris.com/booksearch?keyword='+x)
        browser.visit(url)
        for y in range(1):
            # HTML object
            html = browser.html
            # Parse HTML with Beautiful Soup
            soup = BeautifulSoup(html, 'html.parser')
            # Retrieve all elements that contain book information
            price = soup.find('meta',itemprop='price')['content']
        
            print(price)
            # Iterate through each book
            if price != 0:
                alibris_pandas_df = alibris_pandas_df.append(pd.DataFrame({'ISBN':x,'Alibris Price': price}, index=[0]), ignore_index=True)
    except TypeError:
        pass
            
```
